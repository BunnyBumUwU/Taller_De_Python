Listas

El programa calcula el promedio de una lista de calificaciones usando dos ciclos distintos y muestra los resultados parciales en cada iteración.

```python
calif = [4, 6, 9, 7, 5]

suma = 0
for i in range(len(calif)):
  #print(f'Posicion {i} Calificaion): {{calif[i]}')
  suma = suma + calif[i]

  prom = suma / len(calif)
  print(prom)

  op = 0
  for  e in calif:
    op += e

    print(f'Promedio: {op/len(calif)}')
```

Actividad 2: Dada una secuencia de 15 numeros, imprimir la cantidad de numeros pares, converti los numeros impares en pares, e imprimir la lista completa

```python
import random
numeros = []
for _ in range(15):
  numeros.append(random.randint(1, 100)) # Generando números entre 1 y 100

print(f"Secuencia original: {numeros}")

contador_pares = 0
for num in numeros:
  if num % 2 == 0:
    contador_pares += 1

print(f"Cantidad de números pares en la secuencia original: {contador_pares}")

numeros_modificados = []
for num in numeros:
  if num % 2 != 0:
    numeros_modificados.append(num + 1) # Convertir impar a par sumando 1
  else:
    numeros_modificados.append(num)

# 4. Imprimir la lista completa (modificada)
print(f"Secuencia con números impares convertidos a pares: {numeros_modificados}")
```

El programa calcula el promedio de tres parciales para cada registro, asigna un estatus de aprobado o reprobado y guarda los resultados en un nuevo archivo Excel.

```python
import pandas as pd
import numpy as np

df = pd.read_excel('datos.xlsx')
df.head()

df['prom'] = df[['Parcial 1', 'Parcial 2', 'Parcial 3']].mean(axis=1)
df['estatus'] = np.where(df['prom'] >= 7, 'Aprobado', 'Reprobado')

df.to_excel('datos_mod.xlsx', index=False)
```

Grafica

El programa grafica cómo cambia la temperatura con el tiempo usando Matplotlib y muestra la gráfica en pantalla.

```python

import matplotlib.pyplot as plt

tiempo = [1,4,6,8,10,12]
temperatura = [20,23,27,25,23,19]

plt.plot(tiempo, temperatura)
plt.xlabel('Tiempo (hrs)')
plt.ylabel('Temperatura (°C)')
plt.grid(True)
plt.show()
```
